```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = 50
P = int(input("Enter the premium: "))
S_range = int(input("Enter the Strike Price: "))
```

```python
S = np.linspace(K-S_range, K+S_range, 200)
```

```python
def calculate_payoffs(S, K, P):
    return {
        'Buy Call': np.maximum(S - K, 0) - P,  
        'Sell Call': P - np.maximum(S - K, 0),
        'Buy Put': np.maximum(K - S, 0) - P,
        'Sell Put': P - np.maximum(K - S, 0)
    }
```

```python
payoffs = calculate_payoffs(S, K, P)

plt.figure(figsize=(14, 10))
colors = ['#2E8B57', '#B22222', '#4169E1', '#FF8C00']
```

```python
for i, (title, data) in enumerate(payoffs.items(), 1):
    plt.subplot(2, 2, i)
    plt.plot(S, data, color=colors[i-1], linewidth=2)
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
    plt.axvline(K, color='gray', linestyle=':', linewidth=1)
    plt.title(f"{title} Payoff", fontweight='bold')
    plt.xlabel('Spot Price (S)')
    plt.ylabel('Profit/Loss')
    plt.grid(alpha=0.3)
    plt.fill_between(S, data, where=(data >= 0), 
                     color=colors[i-1], alpha=0.1)

plt.tight_layout()
plt.show()
```
